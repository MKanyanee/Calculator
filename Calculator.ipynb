# I. Using win32com for talking part

Note: This program only speak the result.

```python
from win32com.client import Dispatch
speak = Dispatch("SAPI.SpVoice").Speak

from tkinter import * #use * to import everything from tkinter

## --- Use tkinter to create window and buttons -----------------
win = Tk() # create pop up window
win.title("Unicorn Calculator") # window's title
win.geometry('550x600') # window's size
win['bg']="Powder Blue" # window's background
win.resizable(1,1)   
# if want to prevent from resizing the window, use win.resizable(0, 0)

##--- Functions -------------------------------------------------
# 'button_click' function 
# updates the input field when entering a number
def button_click(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)  

    
# 'button_clear' function 
# clear the input field
def button_clear(): 
    global expression 
    expression = "" 
    input_text.set("")


# 'button_equal' calculates 
# the expression in the input field
def button_equal():
    global expression
    result = str(eval(expression)) # 'eval':This function is used to evaluates the string expression directly
    input_text.set(result)
    expression = "" 
    speak(result)
    
expression = ""


##--- CALCULATOR'S BUTTONS --------------------------------------
# Unicorn colors:
'''
upurple = "#c9b1ff"
upink = "#ffcaf2"
ured = "ffb2b1"  
uyellow = "#fff3ad"
ugreen = "#bcffbc"
ublue = "#2aedff"
'''

# 'StringVar()' :It is used to get the instance of input field
input_text = StringVar()
 
# create a frame for the input field
input_frame = Frame(win, width=40, height=50, bd=0, bg = "#2aedff", highlightbackground="#ffcaf2",  highlightthickness=5) #highlightcolor="#ffcaf2",
 
input_frame.pack(side=TOP)
 
# create an input field inside the 'Frame'
input_field = Entry(input_frame, font=('curlz mt', 18, 'bold'), textvariable=input_text, width=34, bg="#bcffbc", bd=0, justify=RIGHT)
 
input_field.grid(row=0, column=0)
 
input_field.pack(ipady=10) # 'ipady' is internal padding to increase the height of input field
 
# create another 'Frame' for the button below the 'input_frame'
buttons_frame = Frame(win, width=350, height=300, bd=5, bg="#fff3ad") #,  highlightbackground="#fff3ad", highlightthickness=5)
 
buttons_frame.pack()
 
# first row
 
clear = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "CLEAR", fg = "white", width = 32, height = 3, bd = 0, bg = "#c9b1ff", cursor = "hand2", command = lambda: button_clear()).grid(row = 0, column = 0, columnspan = 3, padx = 1, pady = 1)
 
divide = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "/", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("/")).grid(row = 0, column = 3, padx = 1, pady = 1)
 
# second row
 
seven = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "7", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(7)).grid(row = 1, column = 0, padx = 1, pady = 1)
 
eight = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "8", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(8)).grid(row = 1, column = 1, padx = 1, pady = 1)
 
nine = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "9", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(9)).grid(row = 1, column = 2, padx = 1, pady = 1)
 
multiply = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "*", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("*")).grid(row = 1, column = 3, padx = 1, pady = 1)
 
# third row
 
four = Button(buttons_frame,  font=('curlz mt', 15, 'bold'), text = "4", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(4)).grid(row = 2, column = 0, padx = 1, pady = 1)
 
five = Button(buttons_frame,  font=('curlz mt', 15, 'bold'), text = "5", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(5)).grid(row = 2, column = 1, padx = 1, pady = 1)
 
six = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "6", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(6)).grid(row = 2, column = 2, padx = 1, pady = 1)
 
minus = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "-", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("-")).grid(row = 2, column = 3, padx = 1, pady = 1)
 
# fourth row
 
one = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "1", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(1)).grid(row = 3, column = 0, padx = 1, pady = 1)
 
two = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "2", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(2)).grid(row = 3, column = 1, padx = 1, pady = 1)
 
three = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "3", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(3)).grid(row = 3, column = 2, padx = 1, pady = 1)
 
plus = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "+", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("+")).grid(row = 3, column = 3, padx = 1, pady = 1)
 
# fourth row
zero = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "0", fg = "#2aedff", width = 21, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(0)).grid(row = 4, column = 0, columnspan = 2, padx = 1, pady = 1)
 
point = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = ".", fg = "white", width = 10, height = 3, bd = 0, bg = "#c9b1ff", cursor = "hand2", command = lambda: button_click(".")).grid(row = 4, column = 2, padx = 1, pady = 1)
 
equals = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "=", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: {speak("equal"), button_equal()}).grid(row = 4, column = 3, padx = 1, pady = 1)
 
win.mainloop()
```

# II. Using ppyttsx3 for speaking function.

Note: this program speaks all input.

```python
from tkinter import * #use * to import everything from tkinter
import pyttsx3

## --- Use tkinter to create window and buttons -----------------
win = Tk() # create pop up window
win.title("Unicorn Calculator") # window's title
win.geometry('550x600') # window's size
win['bg']="Powder Blue" # window's background
win.resizable(1,1)   
# if want to prevent from resizing the window, use win.resizable(0, 0)

##--- Functions -------------------------------------------------
# 'button_click' function 
# updates the input field when entering a number
def button_click(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)
    engine = pyttsx3.init()
    if item == "/":    
        engine.say("devide by")   
    elif item == "*":
        engine.say("multiply by")    
    elif item == "-":
        engine.say("minus") 
    elif item == ".":
        engine.say("point")
    else:
        engine.say(item)   
    engine.runAndWait()   

    
# 'button_clear' function 
# clear the input field
def button_clear(): 
    global expression 
    expression = "" 
    input_text.set("")
    engine = pyttsx3.init()
    engine.say("clear")   
    engine.runAndWait() 

# 'button_equal' calculates 
# the expression in the input field
def button_equal():
    global expression
    result = str(eval(expression)) # 'eval':This function is used to evaluates the string expression directly
    input_text.set(result)
    expression = ""
    engine = pyttsx3.init()
    engine.say("equal")
    engine.say(result)   
    engine.runAndWait()        
    
expression = ""


##--- CALCULATOR'S BUTTONS --------------------------------------
# Unicorn colors:
# purple #c9b1ff  # pink   #ffcaf2
# red    #ffb2b1  # yellow #fff3ad
# green  #bcffbc  # blue   #2aedff
upurple = "#c9b1ff"
upink = "#ffcaf2"
ured = "ffb2b1"  
uyellow = "#fff3ad"
ugreen = "#bcffbc"
ublue = "#2aedff"

# 'StringVar()' :It is used to get the instance of input field
input_text = StringVar()
 
# create a frame for the input field
input_frame = Frame(win, width=40, height=50, bd=0, bg = "#2aedff", highlightbackground="#ffcaf2",  highlightthickness=5) #highlightcolor="#ffcaf2",
 
input_frame.pack(side=TOP)
 
# create an input field inside the 'Frame'
input_field = Entry(input_frame, font=('curlz mt', 18, 'bold'), textvariable=input_text, width=34, bg="#bcffbc", bd=0, justify=RIGHT)
 
input_field.grid(row=0, column=0)
 
input_field.pack(ipady=10) # 'ipady' is internal padding to increase the height of input field
 
# create another 'Frame' for the button below the 'input_frame'
buttons_frame = Frame(win, width=350, height=300, bd=5, bg="#fff3ad") #,  highlightbackground="#fff3ad", highlightthickness=5)
 
buttons_frame.pack()
 
# first row
 
clear = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "CLEAR", fg = "white", width = 32, height = 3, bd = 0, bg = "#c9b1ff", cursor = "hand2", command = lambda: button_clear()).grid(row = 0, column = 0, columnspan = 3, padx = 1, pady = 1)
 
divide = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "/", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("/")).grid(row = 0, column = 3, padx = 1, pady = 1)
 
# second row
 
seven = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "7", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(7)).grid(row = 1, column = 0, padx = 1, pady = 1)
 
eight = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "8", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(8)).grid(row = 1, column = 1, padx = 1, pady = 1)
 
nine = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "9", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(9)).grid(row = 1, column = 2, padx = 1, pady = 1)
 
multiply = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "*", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("*")).grid(row = 1, column = 3, padx = 1, pady = 1)
 
# third row
 
four = Button(buttons_frame,  font=('curlz mt', 15, 'bold'), text = "4", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(4)).grid(row = 2, column = 0, padx = 1, pady = 1)
 
five = Button(buttons_frame,  font=('curlz mt', 15, 'bold'), text = "5", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(5)).grid(row = 2, column = 1, padx = 1, pady = 1)
 
six = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "6", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(6)).grid(row = 2, column = 2, padx = 1, pady = 1)
 
minus = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "-", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("-")).grid(row = 2, column = 3, padx = 1, pady = 1)
 
# fourth row
 
one = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "1", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(1)).grid(row = 3, column = 0, padx = 1, pady = 1)
 
two = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "2", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(2)).grid(row = 3, column = 1, padx = 1, pady = 1)
 
three = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "3", fg = "#2aedff", width = 10, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(3)).grid(row = 3, column = 2, padx = 1, pady = 1)
 
plus = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "+", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_click("+")).grid(row = 3, column = 3, padx = 1, pady = 1)
 
# fourth row
zero = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "0", fg = "#2aedff", width = 21, height = 3, bd = 0, bg = "#fff", cursor = "hand2", command = lambda: button_click(0)).grid(row = 4, column = 0, columnspan = 2, padx = 1, pady = 1)
 
point = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = ".", fg = "white", width = 10, height = 3, bd = 0, bg = upurple, cursor = "hand2", command = lambda: button_click(".")).grid(row = 4, column = 2, padx = 1, pady = 1)
 
equals = Button(buttons_frame, font=('curlz mt', 15, 'bold'), text = "=", fg = "white", width = 10, height = 3, bd = 0, bg = "#2aedff", cursor = "hand2", command = lambda: button_equal()).grid(row = 4, column = 3, padx = 1, pady = 1)
 
win.mainloop()
```
